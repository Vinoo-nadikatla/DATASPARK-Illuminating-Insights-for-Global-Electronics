```python
import re
from datetime import datetime
# Function to calculate age
def calculate_age(birthdate):
    if pd.isna(birthdate):
        return None
    today = datetime.today()
    return today.year - birthdate.year - ((today.month, today.day) < (birthdate.month, birthdate.day))

# Function to extract numeric value from string
def extract_numeric(value):
    numeric_string = re.sub(r'[^\d.]', '', value)
    return float(numeric_string) if numeric_string else 0.0
```

```python
df_customer.columns = df_customer.columns.str.lower().str.replace(" ", "_")
df_customer['birthday'] = pd.to_datetime(df_customer['birthday'], errors='coerce')
df_customer['age'] = df_customer['birthday'].apply(calculate_age)

```

```python
df_customer['zip_code'] = pd.to_numeric(df_customer['zip_code'], errors='coerce')
```

```python
df_sales = pd.read_csv(r"C:\Users\rames\Downloads\Sales.csv")
df_sales
```

```python
df_sales.columns=df_sales.columns.str.lower().str.replace(" ", "_")
```

```python
df_sales['order_date'] = pd.to_datetime(df_sales['order_date'], errors='coerce')
df_sales['delivery_date'] = pd.to_datetime(df_sales['delivery_date'], errors='coerce')

# Filter out rows where either 'order_date' or 'delivery_date' is missing
filtered_df = df_sales.dropna(subset=['order_date', 'delivery_date']).copy()

# Calculate the difference in days
filtered_df['date_difference'] = (filtered_df['delivery_date'] - filtered_df['order_date']).dt.days

# Compute the average difference
average_difference = filtered_df['date_difference'].mean()

# Fill missing 'delivery_date' values with 'order_date' + average difference
df_sales['delivery_date'] = df_sales['delivery_date'].fillna(df_sales['order_date'])
df_sales['shipping_days'] = (df_sales['delivery_date'] - df_sales['order_date']).dt.days
```

```python

import pandas as pd
import pymysql
from sqlalchemy import create_engine

# Connection to MySQL server without specifying a database to create it first
connection = pymysql.connect(host='localhost', user='root', passwd='123456789')

try:
    with connection.cursor() as cursor:
        # Create the database if it doesn't exist
        cursor.execute("CREATE DATABASE IF NOT EXISTS Capstone2_Dataspark")
finally:
    connection.close()

# Now connect to the newly created or existing database
engine = create_engine("mysql+pymysql://root:123456789@localhost/Capstone2_Dataspark")

# Dictionary of dataframes and their corresponding table names
dataframes = {
    'merged_DataFrame': merged_df
}

# Loop through each dataframe and write it to the SQL database
for table_name, df in dataframes.items():
    # Convert columns to uppercase if they are of object type
    df = df.apply(lambda x: x.str.upper() if x.dtype == "object" else x)
    
    # Use pandas to_sql to create and insert data into the table
    df.to_sql(table_name, con=engine, if_exists='replace', index=False)
    
    print(f"Table {table_name} created and data inserted successfully!")

# Close the connection (optional, SQLAlchemy handles it)
engine.dispose()
```

```python


#1.To get Customer Segmentation by Gender
SELECT gender, COUNT(DISTINCT customerkey) AS total_customers, SUM(quantity * price_usd) AS total_sales
FROM merged_dataframe
GROUP BY gender
ORDER BY total_sales DESC;

#2.To get top 5 most profitable products
SELECT product_name, 
    SUM((price_usd - cost_usd) * quantity) AS total_profit
FROM merged_dataframe
GROUP BY product_name
ORDER BY total_profit DESC
LIMIT 5;

#3.To get Total Sales Over Time
SELECT DATE_FORMAT(order_date, '%Y-%m') AS sales_month, SUM(quantity * price_usd) AS total_sales
FROM merged_dataframe
GROUP BY sales_month
ORDER BY sales_month ASC;

#4.To get Best Performing Stores
SELECT storekey, country_store, SUM(quantity * price_usd) AS total_sales
FROM merged_dataframe
GROUP BY storekey, country_store
ORDER BY total_sales DESC
LIMIT 10;
 
#5. To get Top 10 Customers by Spending
SELECT customerkey, name, SUM(quantity * price_usd) AS total_spent
FROM merged_dataframe
GROUP BY customerkey, name
ORDER BY total_spent DESC
LIMIT 10;


#6.To get Sales by Product Category
SELECT category, SUM(quantity * price_usd) AS total_sales
FROM merged_dataframe
GROUP BY category
ORDER BY total_sales DESC;

#7. To get Average Shipping Days per Order
SELECT order_number, AVG(shipping_days) AS average_shipping_days
FROM merged_dataframe
GROUP BY order_number
ORDER BY average_shipping_days DESC;

#8.To get Product Sales Distribution by Country
SELECT country, product_name, SUM(quantity) AS total_quantity_sold, SUM(price_usd * quantity) AS total_sales
FROM merged_dataframe
GROUP BY country, product_name
ORDER BY country, total_sales DESC;

#9.To get Impact of Exchange Rates on Sales
SELECT currency_code, SUM(price_usd * quantity * exchange) AS adjusted_sales_usd
FROM merged_dataframe
GROUP BY currency_code
ORDER BY adjusted_sales_usd DESC;

#10.To get Store Performance Based on Operational Years
SELECT storekey, country_store, FLOOR(DATEDIFF(CURDATE(), open_date) / 365) AS years_open, SUM(price_usd * quantity) AS total_sales
FROM merged_dataframe
GROUP BY storekey, country_store, years_open
ORDER BY total_sales DESC;
```
